```python
# login_ig(): 登入 IG 正式帳戶，取得 CST + TOKEN
import requests

def login_ig(api_key, username, password):
    url = "https://api.ig.com/gateway/deal/session"
    headers = {
        "X-IG-API-KEY": api_key,
        "Content-Type": "application/json",
        "Accept": "application/json"
    }
    payload = {
        "identifier": username,
        "password": password
    }

    response = requests.post(url, headers=headers, json=payload)

    if response.status_code == 200:
        cst = response.headers.get("CST")
        token = response.headers.get("X-SECURITY-TOKEN")
        print("✅ 登入成功")
        print("CST:", cst)
        print("TOKEN:", token)
        return cst, token
    else:
        print("❌ 登入失敗", response.status_code, response.text)
        return None, None
```

```python
import requests
import json

# 登入 IG 正式帳戶，取得 CST + TOKEN
def login_ig(api_key, username, password):
    url = "https://api.ig.com/gateway/deal/session"
    headers = {
        "X-IG-API-KEY": api_key,
        "Content-Type": "application/json",
        "Accept": "application/json"
    }
    payload = {
        "identifier": username,
        "password": password
    }

    response = requests.post(url, headers=headers, json=payload)

    if response.status_code == 200:
        cst = response.headers.get("CST")
        x_token = response.headers.get("X-SECURITY-TOKEN")
        print("✅ 登入成功")
        print("CST:", cst)
        print("TOKEN:", x_token)
        return cst, x_token
    else:
        print("❌ 登入失敗:", response.status_code, response.text)
        return None, None


# 下單函數 (正式倉，支持 Gold CFD)
def place_order_live(api_key, cst, x_security_token, account_id="QL5HA"):
    url = "https://api.ig.com/gateway/deal/positions/otc"
    headers = {
        "X-IG-API-KEY": api_key,
        "CST": cst,
        "X-SECURITY-TOKEN": x_security_token,
        "Content-Type": "application/json",
        "Version": "2"
    }
    
    payload = {
        "epic": "CS.D.CFDGOLD.CFDGC.IP",
        "expiry": "-",
        "direction": "BUY",
        "size": 0.2,
        "orderType": "MARKET",
        "guaranteedStop": False,
        "forceOpen": True,
        "currencyCode": "USD",
        "accountId": account_id
    }
    
    print("⭕ 正在發送正式倉下單請求...")
    try:
        response = requests.post(url, headers=headers, data=json.dumps(payload))
        if response.status_code == 200:
            data = response.json()
            deal_ref = data.get("dealReference", "")
            print("✅ 正式倉下單成功！參考代碼:", deal_ref)
            return deal_ref
        else:
            print("❌ 正式倉下單失敗，狀態碼:", response.status_code)
            print("⚠ 錯誤內容:", response.text)
            return None
    except requests.exceptions.RequestException as e:
        print(f"❌ 請求異常: {e}")
        return None


# 查詢訂單狀態
def check_order_status(api_key, cst, x_security_token, deal_reference):
    url = f"https://demo-api.ig.com/gateway/deal/confirm/{deal_reference}"
    headers = {
        "X-IG-API-KEY": api_key,
        "CST": cst,
        "X-SECURITY-TOKEN": x_security_token,
        "Accept": "application/json"
    }
    
    print("⏳ 正在查詢訂單狀態...")
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            print("✅ 訂單狀態:", response.json())
        else:
            print("❌ 查詢失敗:", response.status_code)
            print("⚠ 錯誤內容:", response.text)
    except requests.exceptions.RequestException as e:
        print(f"❌ 請求異常: {e}")


# 輸入您的 IG API 資料
api_key = "33012a1647711990919caf823cfaa88765bec64e"
username = "LUCIFINIL"
password = "S22334455s"

# 登入並取得 CST 和 TOKEN
cst, x_token = login_ig(api_key, username, password)

# 如果登入成功，進行下單操作
if cst and x_token:
    # 下單操作
    deal_ref = place_order_live(api_key, cst, x_token)
    # 如果下單成功，查詢訂單狀態
    if deal_ref:
        check_order_status(api_key, cst, x_token, deal_ref)
```
